### Reindexing

````python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
````

````python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e']) #reindexing to a new set of indexes
obj2
````

````python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
````

````python
obj3.reindex(np.arange(6), method="ffill") #forward fill method for reindexing
````

````python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)), 
                     index=['a', 'c', 'd'], 
                     columns=['Ohio', 'Texas', 'California'])
frame
````

````python
frame2 = frame.reindex(index=['a', 'b', 'c', 'd'])
frame2
````

````python
states = ["Texas", "Utah", "California"]
frame.reindex(columns=states) #reindex with a keyword
````

````python
frame.reindex(states, axis="columns") #same way to do ^^ with axis keyword instead
````

````python
frame.loc[["a", "d", "c"], ["California", "Texas"]] #reindexing with loc, changing both rows and columns at once
````
