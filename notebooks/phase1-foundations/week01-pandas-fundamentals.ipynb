### Reindexing

````python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
````

````python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e']) #reindexing to a new set of indexes
obj2
````

````python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
````

````python
obj3.reindex(np.arange(6), method="ffill") #forward fill method for reindexing
````

````python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)), 
                     index=['a', 'c', 'd'], 
                     columns=['Ohio', 'Texas', 'California'])
frame
````

````python
frame2 = frame.reindex(index=['a', 'b', 'c', 'd'])
frame2
````

````python
states = ["Texas", "Utah", "California"]
frame.reindex(columns=states) #reindex with a keyword
````

````python
frame.reindex(states, axis="columns") #same way to do ^^ with axis keyword instead
````

````python
frame.loc[["a", "d", "c"], ["California", "Texas"]] #reindexing with loc, changing both rows and columns at once
````

frame.loc only works if all the new index labels already exist
- reindex adds NaN for missing labels

### Indexing, Selection, and Filtering

````python
obj = pd.Series(np.arange(4.), index=['a', 'b', 'c', 'd'])
obj
````

````python
obj.loc['b'] #loc looks for label first column, not integer position. 
obj.iloc [1] #iloc looks for integer position, not label.
````

````python
obj[2:4] #slicing
````

````python
obj[["b", "a", "d"]]
````

````python
obj [[1, 3]]
````

````python
obj[obj < 2] #boolean indexing
````

````python
obj.loc[["b", "a", "d"]] #preferred to label with loc
````

````python
obj1 = pd.Series([1, 2, 3], index=[2, 0, 1])
obj2 = pd.Series([1, 2, 3], index=["a", "b", "c"])
obj1
````

````python
obj1[[0, 1, 2]] #sorting out the order of the index, not the position of the index.
````

````python
obj2 [[0, 1, 2]] #sorting out doesn't change order for this set
````

````python
obj1. iloc[[0, 1, 2]] #sorting out the position of the index, not the order of the index. iloc is always integer position based.
obj2.iloc[[0, 1, 2]] #using iloc instead to reorder
````

````python
obj2.loc["b":"c"] = 5
obj2
````

##### Arithmetic methods with fill values

````python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), 
                   columns=list("abcd"))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)),
                     columns=list("abcde"))
df2.loc[1, "b"] = np.nan
df1
df2
df1 + df2
````

````python
df1.add(df2, fill_value=0)
````

````python
1 / df1
````

````python
df1.rdiv(1) #same as 1/df1, but with reverse division so that the fill value of 0 is replaced with 1/0 = inf instead of 0/1 = 0
````

````python
df1.reindex(columns=df2.columns, fill_value=0)
````

````python
arr = np.arange(12.).reshape((3, 4))
arr
````

````python
arr[0]
````

````python
arr - arr[0] #subtracting the first row of the array from all rows, broadcasting the subtraction across the array
````

````python
frame = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                        columns=list("bde"),
                        index=["Utah", "Ohio", "Texas", "Oregon"])
series = frame.iloc[0]
frame
````
